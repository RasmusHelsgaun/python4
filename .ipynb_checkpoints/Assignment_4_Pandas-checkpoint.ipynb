## Use pandas dataframe with poverty data
Use this source: https://data.worldbank.org/indicator/SI.POV.DDAY and extract the csv file in jupyter notebook.
With the help of Pandas' DataFrame do the following

```python
url = "http://api.worldbank.org/v2/en/indicator/SI.POV.DDAY?downloadformat=csv"
response = requests.get(url)

#Getting the name of the zip file from the headers
fname = response.headers['Content-Disposition'].split('=')[1]

#Downloading the zip file
if response.ok:  # status_code == 200:
    with open(fname, 'wb') as f:
        f.write(response.content)

#Extracting the csv files
zipfile.ZipFile(fname, 'r').extractall('.')
        
print('Downloaded and extracted {}'.format(fname))
```

glob  is useful in any situation where your program needs to look for a list of files on the filesystem with names matching a pattern. If you need a list of filenames that all have a certain extension, prefix, or any common string in the middle, use glob instead of writing code to scan the directory contents yourself.

```python
file_names = glob('API_SI.POV.DDAY_DS2_en_csv_v2_10474275.csv')
poverty_csv = file_names[0]
poverty_csv
```

```python
#Remove the first lines with skiprows
poverty = pd.read_csv(poverty_csv, skiprows=4)
poverty
```

## 1. Create a line graph showing the development in poverty over time in both Argentina, Ivory cost, USA. (be aware of the missing data in particular years)

```python
years = poverty.iloc[:,4:-1].columns

#gets all data from the given countries and returns the values of the years first and the full dataset secondly
def find_poverty_vals_by_country(list_of_countries):
    boolean_mask = poverty['Country Name'].isin(list_of_countries)
    country_data = poverty[boolean_mask]
    country_vals = country_data.iloc[:, 4:-1].values
    return country_vals, country_data
```

```python
list_of_countries = ['Argentina', 'United States', 'Cote d\'Ivoire']

country_vals, country_data = find_poverty_vals_by_country(list_of_countries)

#get all the countrynames for the legend in the plot below
country_names = []
[country_names.append(country) for country in country_data['Country Name'].values]

country_data
```

```python
plt.figure(figsize=(15,8))
plt.title("Development in poverty over time")
plt.xlabel("Years")
plt.ylabel("Rates")

for row in country_vals:
    plt.plot(years, row, marker='o')

plt.xticks(rotation=45)
plt.legend(country_names)

plt.show()
```

## 2. Find the 10 countries with the highest poverty rate, and plot the rates in a bar chart

```python
#Find the max values of all rows
max_pov = poverty.max(axis=1)

#Sort values and slice top 10
sorted_vals = max_pov.sort_values(ascending=False)
top_ten = sorted_vals[:10]

top_ten_countries = []

#find the countryname and year matching the indexes/values from the top 10 countries
for index in top_ten.index:
    added = False
    country = poverty.iloc[index]
    country_name = poverty.iloc[index]['Country Name']

    for val in top_ten.values:
        for year in years:
            if (country[year] == val) & (added == False):
                top_ten_countries.append(str(country_name + " - " + year))
                added = True
```

```python
plt.figure(figsize=(15,8))
plt.title("Top 10 poverty all time")
plt.xlabel("Country")
plt.ylabel("Poverty rate")

plt.bar(top_ten_countries, top_ten.values)

plt.xticks(rotation=45)

plt.show()
```
